```python
# Disini saya akan menggunakan metode logisticregreassion
# fungsi metode ini adalah memprediksi data dan hasilnya bisa berupa 0 atau 1
# logisticregreassion ada yang multiple dan ada yang biasa, tapi kali ini saya ingin mempelajari yang biasa
# saya ingin memprediksi kemungkinan orang membeli barang kita atau tidak
# disini saya import dulu library nya
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot as plt

df = pd.read_csv("Marketing.csv")
df
```

```python
# saya buang kolom yang tidak diperlukan
df = df.drop(["Location"], axis="columns")
df
```

```python
# disini saya latih data nya
X_train, X_test, y_train, y_test = train_test_split(df[["Age", "Gender","Email Opened", "Email Clicked", "Product page visit", "Discount offered" ]], df["Purchased"], test_size=0.1)
```

```python
X_test
```

```python
# kita gunakan logika ini untuk memprediksi nya
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# kita prediksi hasil uji sampel nya
model.predict(X_test)
```

```python
# kita juga bisa prediksi di luar sampel uji kita
model.predict([[13, 0, 1, 0, 2, 1]])
```
