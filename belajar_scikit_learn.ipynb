```python
# Belajar membuat prediksi menggunakan scikit learn

import pandas as pd
from matplotlib import pyplot as plt
from sklearn import linear_model

df = pd.read_csv("canada.csv")
df
```

```python
# membuat model prediksinya (untuk argumen pertama disertai dengan tanda kurung double)

reg = linear_model.LinearRegression()
reg.fit(df[["year"]], df["per capita income (US$)"])
```

```python
# membuat prediksi nya dan menambahkannya ke dalam dataframe baru agar bisa ditambah ke dataframe lama

years_predict = [[2017], [2018], [2019], [2020]]
predictions = reg.predict(years_predict)

final = pd.DataFrame({"year" : [2017,2018,2019,2020],
                      "per capita income (US$)" : predictions})

df = pd.concat([df, final], ignore_index=True)
df
```

```python
# memvisualisasikan data nya menggunakan matplotlib

plt.scatter(df["year"], df["per capita income (US$)"], c = "green")

# membuat plot satu lagi untuk membuat garis lurus miring
plt.plot(df["year"], reg.predict(df[["year"]]), color = "red")
plt.xlabel("Tahun")
plt.ylabel("Pendapatan")

plt.legend()
plt.show()
```
