```python
import pandas as pd
# Mengimpor library pandas untuk mengolah data berbentuk tabel (DataFrame)

from sklearn.model_selection import train_test_split
# Digunakan untuk membagi data menjadi data latih dan data uji

from sklearn.naive_bayes import GaussianNB
# Mengimpor algoritma Naive Bayes Gaussian untuk klasifikasi


df = pd.read_csv("pelanggan.csv")
# Membaca file pelanggan.csv dan menyimpannya ke dalam variabel df

df
# Menampilkan isi dataset
```

```python
df = df[['Attrition_Flag', 'Customer_Age', 'Dependent_count', 'Education_Level', 'Months_on_book']]
# Memilih kolom-kolom yang diperlukan saja agar data lebih fokus dan sederhana

df
# Menampilkan dataset setelah kolom dipilih
```

```python
from sklearn.preprocessing import LabelEncoder
# Mengimpor LabelEncoder untuk mengubah data kategorik menjadi numerik

le_sex = LabelEncoder()
# Membuat objek encoder

df["ada/tidak"] = le_sex.fit_transform(df["Attrition_Flag"])
# Mengubah nilai Attrition_Flag menjadi angka (encoding)

df["ada/tidak"] = df["ada/tidak"].astype(int)
# Mengubah tipe data menjadi integer


df["Edukasi"] = le_sex.fit_transform(df["Education_Level"])
# Mengubah kategori Education_Level menjadi nilai numerik

df["Edukasi"] = df["Edukasi"].astype(int)
# Mengubah tipe data menjadi integer

df
```

```python
df = df[['Attrition_Flag', "ada/tidak" ,'Customer_Age', 'Dependent_count', 'Education_Level', "Edukasi", 'Months_on_book']]
# Menyusun ulang kolom agar data asli dan data hasil encoding tetap terlihat

df
# Menampilkan dataset akhir
```

```python
x = df[["Customer_Age", "Dependent_count", "Edukasi", "Months_on_book"]]
# Menentukan fitur (variabel independen) yang digunakan untuk pelatihan model

y = df["ada/tidak"]
# Menentukan label/target yang akan diprediksi


X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
# Membagi data menjadi 70% data latih dan 30% data uji

X_test
# Menampilkan data uji (fitur)
```

```python
model = GaussianNB()
# Membuat model Naive Bayes Gaussian

model.fit(X_train,y_train)
# Melatih model menggunakan data latih
```

```python
y_test
# Menampilkan label asli dari data uji
```

```python
model.predict(X_test)
# Melakukan prediksi terhadap data uji
```

```python
model.score(X_test,y_test)
# Menghitung akurasi model berdasarkan data uji
```

```python
model.predict([[34, 3, 5, 50]])
# Melakukan prediksi untuk data pelanggan baru:
# Usia 34 tahun, tanggungan 3, tingkat pendidikan (encoded) 5, lama berlangganan 50 bulan
```
